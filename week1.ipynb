```python


import soundfile as sf
import numpy as np

def readAudio(inputFile):

    x, fs = sf.read(inputFile)

    # Ensure audio is long enough
    if len(x) < 50010:
        raise ValueError("Audio file is too short!")

    return x[50000:50010]


files = [
    "/content/mridangam.wav",
    "/content/multiSines.wav",
    "/content/ocean.wav",
    "/content/sax-phrase-short.wav",
    "/content/organ-C3.wav",
    "/content/sine-1000.wav",
    "/content/sine-200.wav",
    "/content/vibraphone-C6.wav",
    "/content/vignesh.wav",
    "/content/speech-female.wav"
]

for f in files:
    try:
        samples = readAudio(f)
        print(f"\nFile: {f}")
        print(samples)
    except Exception as e:
        print(f"\nError with {f}: {e}")
```

```python


import soundfile as sf
import numpy as np

def minMaxAudio(inputFile):



    x, fs = sf.read(inputFile)


    min_val = np.min(x)
    max_val = np.max(x)

    return (float(min_val), float(max_val))



file_path = "/content/ocean.wav"

try:
    result = minMaxAudio(file_path)
    print("Min and Max values:", result)
except Exception as e:
    print("Error:", e)
```
