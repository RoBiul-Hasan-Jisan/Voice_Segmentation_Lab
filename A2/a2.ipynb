```python

import numpy as np
import matplotlib.pyplot as plt

print("Setup complete!")
print(f"NumPy version: {np.__version__}")
```

```python
#  Generate a sinusoid

def genSine(A, f, phi, fs, t):
    """
    Inputs:
        A (float) = amplitude of the sinusoid
        f (float) = frequency of the sinusoid in Hz
        phi (float) = initial phase of the sinusoid in radians
        fs (float) = sampling frequency of the sinusoid in Hz
        t (float) = duration of the sinusoid (in seconds)
    Output:
        x (numpy array) = The generated sinusoid (using np.cos())
    """
    # Number of samples
    N = int(fs * t)
    
    # Time vector
    n = np.arange(N)
    
    # Generate sinusoid: A * cos(2πf * n/fs + φ)
    x = A * np.cos(2 * np.pi * f * n / fs + phi)
    
    return x

# Test the function
print("Testing genSine function")
print("-" * 50)

# Test with given example
A, f, phi, fs, t = 1.0, 10.0, 1.0, 50.0, 0.1
x = genSine(A, f, phi, fs, t)

print(f"Input: A={A}, f={f}, phi={phi}, fs={fs}, t={t}")
print(f"Output shape: {x.shape}")
print(f"Output array: {x}")
print(f"\nExpected output: [0.54030231, -0.63332387, -0.93171798, 0.05749049, 0.96724906]")

# Check if the output matches the expected values (approximately)
expected = np.array([0.54030231, -0.63332387, -0.93171798, 0.05749049, 0.96724906])
if np.allclose(x, expected, atol=1e-6):
    print("\n Test passed!")
else:
    print("\n Test failed!")
    
# Visualize the sinusoid
if len(x) > 1:
    plt.figure(figsize=(10, 4))
    plt.plot(x, 'b-o', markersize=8)
    plt.title(f'Sinusoid: A={A}, f={f}Hz, φ={phi} rad, fs={fs}Hz, t={t}s')
    plt.xlabel('Sample index (n)')
    plt.ylabel('Amplitude')
    plt.grid(True, alpha=0.3)
    plt.show()
```

```python
# Generate a complex sinusoid

def genComplexSine(k, N):
    """
    Inputs:
        k (integer) = frequency index of the complex sinusoid of the DFT
        N (integer) = length of complex sinusoid in samples
    Output:
        cSine (numpy array) = The generated complex sinusoid (length N)
    """
    n = np.arange(N)
    # Complex sinusoid used in DFT: e^(-j * 2π * k * n / N)
    s = np.exp(-1j * 2 * np.pi * k * n / N)
    return s

# Test the function
print("Testing genComplexSine function")
print("-" * 50)

# Test with given example
N, k = 5, 1
cSine = genComplexSine(k, N)

print(f"Input: N={N}, k={k}")
print(f"Output shape: {cSine.shape}")
print(f"Output array:")
for i, val in enumerate(cSine):
    print(f"  cSine[{i}] = {val}")
    
print(f"\nExpected output:")
expected = np.array([1.0 + 0.j, 0.30901699 - 0.95105652j, -0.80901699 - 0.58778525j, 
                     -0.80901699 + 0.58778525j, 0.30901699 + 0.95105652j])

# Check if the output matches expected values
if np.allclose(cSine, expected, atol=1e-6):
    print("\n Test passed!")
else:
    print("\nTest failed!")

# Visualize the complex sinusoid
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(np.real(cSine), 'b-o', label='Real part')
plt.plot(np.imag(cSine), 'r-o', label='Imag part')
plt.title(f'Complex Sinusoid: k={k}, N={N}')
plt.xlabel('Sample index (n)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(np.real(cSine), np.imag(cSine), 'go-')
plt.title('Complex Plane Plot')
plt.xlabel('Real part')
plt.ylabel('Imag part')
plt.axis('equal')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Implement DFT

def DFT(x):
    """
    Input:
        x (numpy array) = input sequence of length N
    Output:
        X (numpy array) = The N point DFT of the input sequence x
    """
    N = x.size
    X = np.array([], dtype=complex)
    
    for k in range(N):
        # Complex exponential e^(j * 2π * k * n / N)
        s = np.exp(1j * 2 * np.pi * k * np.arange(N) / N)
        # DFT formula: X[k] = Σ x[n] * e^(-j * 2π * k * n / N)
        # Using conjugate of s gives e^(-j * 2π * k * n / N)
        X = np.append(X, np.sum(x * np.conjugate(s)))
    
    return X

# Test the function
print("Testing DFT function")
print("-" * 50)

# Test with given example
x = np.array([1, 2, 3, 4])
X = DFT(x)

print(f"Input: x = {x}")
print(f"Output shape: {X.shape}")
print(f"Output: {X}")
print(f"\nExpected: [10.0 + 0.0j, -2.0 + 2.0j, -2.0 - 9.79717439e-16j, -2.0 - 2.0j]")

# Check if the output matches expected values
expected = np.array([10.0 + 0.0j, -2.0 + 2.0j, -2.0 + 0j, -2.0 - 2.0j])
if np.allclose(X, expected, atol=1e-10):
    print("\n Test passed!")
else:
    print("\n Test failed! (Numerical differences due to floating point precision)")

# Test with a simple sinusoid
n = np.arange(64)
x_sine = np.cos(2 * np.pi * 5 * n / 64)  # 5 cycles in 64 samples
X_sine = DFT(x_sine)

# Plot magnitude spectrum
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(x_sine)
plt.title('Input Signal: Cosine wave (5 cycles)')
plt.xlabel('Sample index')
plt.ylabel('Amplitude')
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.stem(np.abs(X_sine))
plt.title('DFT Magnitude Spectrum')
plt.xlabel('Frequency index (k)')
plt.ylabel('Magnitude')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Implement IDFT

def IDFT(X):
    """
    Input:
        X (numpy array) = frequency spectrum (length N)
    Output:
        x (numpy array) = The N point IDFT of the frequency spectrum X
    """
    N = X.size
    n = np.arange(N)
    k = np.arange(N)
    x = np.array([], dtype=complex)
    
    for n_idx in n:
        # Complex exponential: e^(j * 2π * n * k / N)
        s = np.exp(1j * 2 * np.pi * n_idx * k / N)
        # IDFT formula: x[n] = (1/N) * Σ X[k] * e^(j * 2π * k * n / N)
        x = np.append(x, (1.0 / N) * np.sum(X * s))
    
    return x

# Test the function
print("Testing IDFT function")
print("-" * 50)

# Test with given example
X = np.array([1, 1, 1, 1], dtype=complex)
x = IDFT(X)

print(f"Input: X = {X}")
print(f"Output: {x}")
print(f"\nExpected: [1, 0, 0, 0] (with small numerical errors)")

# Check if the output is approximately [1, 0, 0, 0]
expected = np.array([1, 0, 0, 0], dtype=complex)
if np.allclose(np.real(x), np.real(expected), atol=1e-10):
    print("\ Test passed! (IDFT produces expected result)")
else:
    print("\n Test failed!")

# Test DFT-IDFT consistency
print("\n" + "="*50)
print("Testing DFT-IDFT consistency")
print("-" * 50)

# Original signal
x_original = np.array([1, 2, 3, 4, 5, 6, 7, 8])

# Compute DFT
X_dft = DFT(x_original)

# Compute IDFT
x_reconstructed = IDFT(X_dft)

# Compare
print(f"Original signal: {x_original}")
print(f"Reconstructed:   {np.real(x_reconstructed)}")

if np.allclose(x_original, np.real(x_reconstructed), atol=1e-10):
    print("\ Consistency test passed! IDFT(DFT(x)) = x")
else:
    print("\n Consistency test failed!")
```

```python
#  ACompute magnitude spectrum 

# Import the DFT function from Part 3
# Note: Make sure A2Part3.py is in the same directory
# For this notebook, we'll use the DFT function defined above

def genMagSpec(x):
    """
    Input:
        x (numpy array) = input sequence of length N
    Output:
        magX (numpy array) = The magnitude spectrum of the input sequence x (length N)
    """
    # Compute DFT using the DFT function from Part 3
    X = DFT(x)
    
    # Compute magnitude spectrum
    magX = np.abs(X)
    
    return magX

# Test the function
print("Testing genMagSpec function")
print("-" * 50)

# Test with given example
x = np.array([1, 2, 3, 4])
magX = genMagSpec(x)

print(f"Input: x = {x}")
print(f"Output magnitude spectrum: {magX}")
print(f"\nExpected: [10.0, 2.82842712, 2.0, 2.82842712]")

# Check if the output matches expected values
expected = np.array([10.0, 2.82842712, 2.0, 2.82842712])
if np.allclose(magX, expected, atol=1e-6):
    print("\n Test passed!")
else:
    print("\n Test failed!")

# Visualize magnitude spectrum for different signals

print("Visualizing magnitude spectra")


# Test with different signals
test_signals = {
    'Impulse': np.array([1, 0, 0, 0, 0, 0, 0, 0]),
    'Square wave': np.array([1, 1, 1, 1, -1, -1, -1, -1]),
    'Sinusoid': np.cos(2 * np.pi * 2 * np.arange(16) / 16),
    'Random noise': np.random.randn(16)
}

fig, axes = plt.subplots(2, 4, figsize=(15, 6))
fig.suptitle('Magnitude Spectra of Different Signals', fontsize=14)

for idx, (name, signal) in enumerate(test_signals.items()):
    # Plot signal
    ax1 = axes[0, idx]
    ax1.plot(signal, 'b-o', markersize=4)
    ax1.set_title(f'{name} - Signal')
    ax1.set_xlabel('Sample')
    ax1.set_ylabel('Amplitude')
    ax1.grid(True, alpha=0.3)
    
    # Plot magnitude spectrum
    ax2 = axes[1, idx]
    mag_spec = genMagSpec(signal)
    ax2.stem(mag_spec)
    ax2.set_title(f'{name} - Magnitude Spectrum')
    ax2.set_xlabel('Frequency index (k)')
    ax2.set_ylabel('Magnitude')
    ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
