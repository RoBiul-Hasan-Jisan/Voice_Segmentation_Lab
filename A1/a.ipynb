```python
# Using scipy.io.wavfile instead of utilFunctions
import numpy as np
import os
from scipy.io import wavfile

# Set the audio folder path
AUDIO_FOLDER = r'D:\lern voice\Voice_Segmentation_Lab\sounds'

print(f"Audio folder: {AUDIO_FOLDER}")
print(f"Folder exists: {os.path.exists(AUDIO_FOLDER)}")
print("\nSetup complete!")
```

```python
# Cell 2: Define custom wavread and wavwrite functions (replacement for utilFunctions)

def wavread(inputFile):
    """
    Read a wav file and return samples as float32 in range [-1, 1]
    
    Args:
        inputFile: path to wav file
        
    Returns:
        fs: sampling rate
        x: audio samples as float32 array in range [-1, 1]
    """
    fs, x = wavfile.read(inputFile)
    
    # Convert to float32 and normalize to [-1, 1] range
    if x.dtype == np.int16:
        x = x.astype(np.float32) / 32768.0
    elif x.dtype == np.int32:
        x = x.astype(np.float32) / 2147483648.0
    elif x.dtype == np.uint8:
        x = (x.astype(np.float32) - 128) / 128.0
    
    # Ensure mono (take first channel if stereo)
    if len(x.shape) > 1:
        x = x[:, 0]
    
    return fs, x

def wavwrite(y, fs, outputFile):
    """
    Write audio samples to a wav file
    
    Args:
        y: audio samples as float32 array in range [-1, 1]
        fs: sampling rate
        outputFile: path to output wav file
    """
    # Convert from float32 to int16
    y_int16 = (y * 32767).astype(np.int16)
    wavfile.write(outputFile, int(fs), y_int16)

print("Custom wavread and wavwrite functions defined successfully!")
```

```python
#  List all available audio files in the folder
def list_audio_files(folder_path):
    """List all wav files in the specified folder"""
    if os.path.exists(folder_path):
        wav_files = [f for f in os.listdir(folder_path) if f.endswith('.wav')]
        print(f"\nFound {len(wav_files)} audio files in {folder_path}:")
        print("-" * 60)
        for i, file in enumerate(wav_files, 1):
            # Get file size
            file_path = os.path.join(folder_path, file)
            size_bytes = os.path.getsize(file_path)
            size_kb = size_bytes / 1024
            print(f"  {i:2}. {file:<30} ({size_kb:.1f} KB)")
        return wav_files
    else:
        print(f"ERROR: Folder not found: {folder_path}")
        print("Please check the path and try again.")
        return []

# List all available audio files
available_files = list_audio_files(AUDIO_FOLDER)
```

```python
# A1-Part-1 - Reading an audio file (3 points)

def readAudio(inputFile):
    """
    Input:
        inputFile: the path to the wav file      
    Output:
        The function should return a numpy array that contains 10 samples of the audio,
        starting from the 50001th sample to the 50010th sample (10 samples).
    """
    # Read the audio file
    (fs, x) = wavread(inputFile)
    
    # Extract 10 samples starting from index 50000 (50001th sample)
    # Remember: Python indexing starts at 0, so sample 50001 is at index 50000
    y = x[50000:50010]
    
    return y

# Test the function with piano.wav from your folder
piano_file = os.path.join(AUDIO_FOLDER, 'piano.wav')

if os.path.exists(piano_file):
    print("Testing Part 1 with piano.wav")
    print("-" * 50)
    result_part1 = readAudio(piano_file)
    print("First 10 samples from 50001th position:")
    print(result_part1)
    print(f"\nData type: {result_part1.dtype}")
    print(f"Number of samples: {len(result_part1)}")
    print(f"\nExpected output: [-0.06213569, -0.04541154, -0.02734458, -0.0093997, 0.00769066, 0.02319407, 0.03503525, 0.04309214, 0.04626606, 0.0441908]")
else:
    print(f"ERROR: piano.wav not found in {AUDIO_FOLDER}")
    print("Available files:", available_files)
```

```python
#  A1-Part-2 - Basic operations with audio (3 points)

def minMaxAudio(inputFile):
    """
    Input:
        inputFile: file name of the wav file (including path)
    Output:
        A tuple of the minimum and the maximum value of the audio samples, like: (min_val, max_val)
    """
    # Read the audio file
    (fs, x) = wavread(inputFile)
    
    # Return min and max as a tuple
    return (np.min(x), np.max(x))

# Test the function with oboe-A4.wav from your folder
oboe_file = os.path.join(AUDIO_FOLDER, 'oboe-A4.wav')

if os.path.exists(oboe_file):
    print("Testing Part 2 with oboe-A4.wav")
    print("-" * 50)
    min_val, max_val = minMaxAudio(oboe_file)
    print(f"Minimum value: {min_val:.8f}")
    print(f"Maximum value: {max_val:.8f}")
    print(f"Tuple: ({min_val:.8f}, {max_val:.8f})")
    print(f"\nExpected output: (-0.83486432, 0.56501967)")
else:
    print(f"ERROR: oboe-A4.wav not found in {AUDIO_FOLDER}")
    print("Available files:", available_files)
```

```python
#  A1-Part-3 - Python array indexing (4 points)

def hopSamples(x, M):
    """
    Inputs:
        x: input numpy array
        M: hop size (positive integer)
    Output:
        A numpy array containing every Mth element in x, starting from the first element in x.
    """
    # Using slicing with step M, starting from the first element (index 0)
    return x[0::M]

# Test the function
x_test = np.arange(10)
M_test = 2
result_part3 = hopSamples(x_test, M_test)

print("Testing Part 3 - hopSamples function")
print("-" * 50)
print(f"Input array: {x_test}")
print(f"Hop size M = {M_test}")
print(f"Output: {result_part3}")
print(f"Expected output: [0, 2, 4, 6, 8]")
print(f"Correct: {np.array_equal(result_part3, np.array([0, 2, 4, 6, 8]))}")
```

```python
# A1-Part-4 - Downsampling audio 

def downsampleAudio(inputFile, M):
    """
    Inputs:
        inputFile: file name of the wav file (including path)
        M: downsampling factor (positive integer)
    """
    # Read the original audio file
    (fs, x) = wavread(inputFile)
    
    # Downsample by taking every Mth sample using hopSamples function
    y = hopSamples(x, M)
    
    # Calculate the new sampling rate
    new_fs = fs / M
    
    # Create output filename in the same folder
    base_name = os.path.basename(inputFile)
    name_without_ext = os.path.splitext(base_name)[0]
    output_file = os.path.join(AUDIO_FOLDER, name_without_ext + '_downsampled.wav')
    
    # Save the downsampled audio
    wavwrite(y, new_fs, output_file)
    
    print(f"Original sampling rate: {fs} Hz")
    print(f"Downsampling factor: M = {M}")
    print(f"New sampling rate: {new_fs} Hz")
    print(f"Original number of samples: {len(x)}")
    print(f"New number of samples: {len(y)}")
    print(f"Output file saved as: {output_file}")
    
    return y, new_fs

# Test downsampling if vibraphone-C6.wav exists
vibraphone_file = os.path.join(AUDIO_FOLDER, 'vibraphone-C6.wav')

if os.path.exists(vibraphone_file):
    print("Testing Part 4 (Optional) - Downsampling Audio")
    print("-" * 50)
    M = 16
    downsampled_audio, new_sr = downsampleAudio(vibraphone_file, M)
    
```
