```python
import librosa
import matplotlib.pyplot as plt

audio, sr = librosa.load("audio.wav")  # audio = samples, sr = sampling rate
print(f"Audio duration: {len(audio)/sr:.2f} seconds")
```

- audio is a NumPy array of amplitude values.
- sr is the sampling rate (usually 22050 Hz by default in Librosa).

```python
import numpy as np

times = np.arange(len(audio)) / sr  # convert sample index to time
plt.figure(figsize=(14,4))
plt.plot(times, audio)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Audio Waveform")
plt.show()
```

```python
import librosa.display

plt.figure(figsize=(14, 5))
D = librosa.amplitude_to_db(np.abs(librosa.stft(audio)), ref=np.max)
librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram (dB)')
plt.show()
```

```python
mfccs = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=13)
plt.figure(figsize=(14,5))
librosa.display.specshow(mfccs, sr=sr, x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.show()
```
